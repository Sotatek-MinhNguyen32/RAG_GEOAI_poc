# Test MCP Search
Notebook đơn giản để test MCP controller search pipeline.

```python
import sys, json
sys.path.insert(0, "..")

from mcp.controller import handle_query
```

```python
async def search(query: str, top_k: int = 5) -> dict:
    """Search qua MCP controller."""
    result = await handle_query({"query": query, "top_k": top_k})
    print(json.dumps(result, indent=2, ensure_ascii=False, default=str))
    return result
```

```python
# Test: tìm ảnh "rừng núi"
result = await search("rừng núi")
```

```python
# Test thêm query khác
result2 = await search("satellite image of urban area")
```

```python
# Xem chi tiết kết quả đầu tiên
if result["results"]:
    top = result["results"][0]
    print(f"Image: {top['image_id']}")
    print(f"Score: {top['score']}")
    print(f"Sources: {top['sources']}")
    print(f"Description: {top.get('description', 'N/A')[:200]}")
```
